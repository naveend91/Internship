C) Contributors count
D) Language used

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException

from selenium.webdriver.support.ui import WebDriverWait
```

```python
####-------------------
# connecting to the web driver
driver=webdriver.Chrome()

# getting the webpage of mentioned url
url = ("https://github.com/")
driver.get(url)
```

```python
search=driver.find_element(By.XPATH,'//input[@id="query-builder-test"]')
search.send_keys('trending repositories')
click=driver.find_element(By.XPATH,'//span[@class="ActionListItem-description QueryBuilder-ListItem-trailing"]')
click.click()
```

```python
# creating empty list
URLs = []
tittle = []

lang = []

# fetching urls for each repository
repository = driver.find_elements(By.XPATH,'//a[@class="Link__StyledLink-sc-14289xe-0 fIqerb"]')
for i in repository:
    URLs.append(i.get_attribute("href"))
 #tittle   


tit=driver.find_elements(By.XPATH,'//span[@class="Text-sc-17v1xeu-0 qaOIC search-match"]')
for i in tit:
    tittle.append(i.text)
#print(tittle)
```

```python
Contributors = []

Language = []

for i in URLs:
    driver.get(i)
    time.sleep(5)

    try:
        for i in driver.find_elements(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]'):# check this loop
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('-')
        
    try:
        desc = driver.find_element(By.XPATH,'//div[@class="BorderGrid-row"][6]/div/h2')
        Contributors.append(desc.text)
    except NoSuchElementException:
        Contributors.append('-')
```

```python
Github = pd.DataFrame({})
Github['Repository Title'] = tittle
Github['Contributors Count'] = Contributors
Github['Language Used'] = Language
Github
```

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common. by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver=webdriver.Chrome()
driver.get(' https:/www.billboard.com/')
charts=driver.find_element(By.LINK_TEXT,'CHARTS')
charts.click()
```

```python
top100=driver.find_element(By.XPATH,'//a[@class="c-label__link lrv-u-color-brand-primary u-color-black@mobile-max a-font-primary-bold lrv-u-font-size-32 u-whitespace-nowrap@tablet"]')
top100.click()
```

```python
#in below code relative xpath use change li number(columns)

#song_name
song_name=[]
song=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"][1]/ul/li/h3')
for i in song:
    song_name.append(i.text)
#len(song_name)

#Artist Name
artist_name=[]
song=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul[1]/li[1]/span')# check this class new learning
for i in song:
    artist_name.append(i.text)
#len(artist_name)

#week on board
week_board=[]
boardweek=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul[1]/li[6]')
for i in boardweek:
    week_board.append(i.text)
#len(week_board)
    
    
#Peak_renk
peak_rank=[]
peak=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[5]')
for i in peak:
    peak_rank.append(i.text)
#len(peak_rank)


#last week rank
last_weekrank=[]
last_week=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul[1]/li[4]')
for i in last_week:
    last_weekrank.append(i.text)
#len(last_weekrank)


df=pd.DataFrame({'Song name':song_name,'Artist name':artist_name,'Last week rank':last_weekrank,'Peak rank':peak_rank,'Weeks on board':week_board})
df
```

## 8 . Details of Datasets from UCI machine learning repositories.   
Url = https://archive.ics.uci.edu/  You 
have to find the following details:   
A) Dataset name   
B) Data type   
C) Task   
D) Attribute type   
E) No of instances   
F) No of attribute G) Year   
Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
datasets=driver.find_element(By.LINK_TEXT,'Datasets')
datasets.click()
```

```python
product_url=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')#the link u want to click
    for i in url:
        product_url.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
    next_button.click()
    time.sleep(3)
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
Noof_instances=[]
Noof_attribute=[]
Year=[]




for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    #Dataset_name
    try:
        brand=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]') 
        Dataset_name.append(brand.text)
    except NoSuchElementException:
        Dataset_name.append('-')
        
        
        
#Data_type        
    try:
        n=driver.find_element(By.XPATH,'//div[@class="col-span-4"][1]/p[1]') 
        Data_type.append(n.text)
    except NoSuchElementException:
        Data_type.append('-')
        
        
        #task        
    try:
        p=driver.find_element(By.XPATH,'//div[@class="col-span-4"][3]/p[1]') 
        Task.append(p.text)
    except NoSuchElementException:
        Task.append('-')
        
        #Attribute_type        
    try:
        r=driver.find_element(By.XPATH,'//div[@class="col-span-4"][4]/p[1]') 
        Attribute_type.append(r.text)
    except NoSuchElementException:
        Attribute_type.append('-')
        
        
#Noof_instances        
    try:
        a=driver.find_element(By.XPATH,'//div[@class="col-span-4"][5]/p[1]') 
        Noof_instances.append(a.text)
    except NoSuchElementException:
        Noof_instances.append('-')
        
        
#Noof_attribute        
    try:
        a=driver.find_element(By.XPATH,'//div[@class="col-span-4"][6]/p[1]') 
        Noof_attribute.append(a.text)
    except NoSuchElementException:
        Noof_attribute.append('-')
        



```

```python
len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(Noof_instances),len(Noof_attribute)
```

```python
df = pd.DataFrame({})
df['Dataset_name'] = Dataset_name
df['Data_type'] = Data_type
df['Task'] = Task
df['Attribute_type'] = Attribute_type
df['Noof_instances'] = Noof_instances
df['Noof_attribute']=  Noof_attribute
df
```
