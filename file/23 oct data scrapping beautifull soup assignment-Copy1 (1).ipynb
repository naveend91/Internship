```python
#ans 3

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape and create a DataFrame
def scrape_icc_rankings(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        data = []

        # Find the appropriate HTML elements for the data
        for row in soup.find_all('tr', class_='table-body'):
            cols = row.find_all('td')
            rank = cols[0].text.strip()
            team = cols[1].text.strip()
            matches = cols[2].text.strip()
            points = cols[3].text.strip()
            rating = cols[4].text.strip()

            data.append([rank, team, matches, points, rating])

        # Create a DataFrame from the scraped data
        df = pd.DataFrame(data, columns=['Rank', 'Team', 'Matches', 'Points', 'Rating'])
        return df
    else:
        print("Failed to retrieve the web page")

# URLs for scraping data
odi_teams_url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
odi_batsmen_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting'
odi_bowlers_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'

# Scrape and create DataFrames for ODI teams, batsmen, and bowlers
odi_teams_df = scrape_icc_rankings(odi_teams_url)
odi_batsmen_df = scrape_icc_rankings(odi_batsmen_url)
odi_bowlers_df = scrape_icc_rankings(odi_bowlers_url)

# Display the DataFrames
print("Top 10 ODI Teams:")
print(odi_teams_df.head(10))

print("\nTop 10 ODI Batsmen:")
print(odi_batsmen_df.head(10))

print("\nTop 10 ODI Bowlers:")
print(odi_bowlers_df.head(10))
```

```python
#ans 4

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape and create a DataFrame
def scrape_icc_rankings(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        data = []

        # Find the appropriate HTML elements for the data
        for row in soup.find_all('tr', class_='table-body'):
            cols = row.find_all('td')
            rank = cols[0].text.strip()
            name = cols[1].text.strip()
            team = cols[2].text.strip()
            rating = cols[3].text.strip()

            data.append([rank, name, team, rating])

        # Create a DataFrame from the scraped data
        df = pd.DataFrame(data, columns=['Rank', 'Name', 'Team', 'Rating'])
        return df
    else:
        print("Failed to retrieve the web page")

# URLs for scraping data
odi_teams_women_url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
odi_batting_women_url = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting'
odi_all_rounder_women_url = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder'

# Scrape and create DataFrames for ODI teams, batting players, and all-rounders
odi_teams_women_df = scrape_icc_rankings(odi_teams_women_url)
odi_batting_women_df = scrape_icc_rankings(odi_batting_women_url)
odi_all_rounder_women_df = scrape_icc_rankings(odi_all_rounder_women_url)

# Display the DataFrames
print("Top 10 Women's ODI Teams:")
print(odi_teams_women_df.head(10))

print("\nTop 10 Women's ODI Batting Players:")
print(odi_batting_women_df.head(10))

print("\nTop 10 Women's ODI All-rounders:")
print(odi_all_rounder_women_df.head(10))
```

```python
#ans 5
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape and create a DataFrame
def scrape_cnbc_news(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        data = []

        # Find the appropriate HTML elements for the data
        articles = soup.find_all('div', class_='Card-Item')
        for article in articles:
            headline = article.find('h3', class_='Card-Headline').text.strip()
            time = article.find('time').text.strip()
            news_link = article.find('a', class_='Card-Link')['href']

            data.append([headline, time, news_link])

        # Create a DataFrame from the scraped data
        df = pd.DataFrame(data, columns=['Headline', 'Time', 'News Link'])
        return df
    else:
        print("Failed to retrieve the web page")

# URL for scraping data
cnbc_world_news_url = 'https://www.cnbc.com/world/?region=world'

# Scrape and create a DataFrame for news details
cnbc_news_df = scrape_cnbc_news(cnbc_world_news_url)

# Display the DataFrame
print(cnbc_news_df)

```
